```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
try:
    source = requests.get('https://www.bestproducts.com/lifestyle/g3486/best-selling-products-on-amazon/#:~:text=The%2070%20Best-Selling%20Products%20on%20Amazon%20Right%20Now,Pet%20Deshedding%20Brush%20Glove.%20...%20More%20items...%20')
    source.raise_for_status()
    soup = BeautifulSoup(source.text, 'html.parser')
    products = soup.find('div', class_="listicle-body-content").find_all('div',
                                                                         class_="listicle-slide listicle-slide-square listicle-slide-product listicle-slide-multi-retailer")

    for product in products:

        rank = product.find('span', class_="listicle-slide-hed-number").text
        name = product.find('span', class_="listicle-slide-hed-text").text
        sold_by = product.find(
            'div', class_="product-slide-details noskim").text
        rate = product.find('div', class_="product-buy-button-wrapper").text

        print(rank, name, sold_by, rate)

except Exception as e:
    print(e)
```
